# 価格・対数価格・収益率・対数収益率

## ドル円

```python
import pandas_datareader.data as pdr
import datetime
start = datetime.datetime(1971,1,4)
end = datetime.datetime(2019,8,30)
JPY_USD = pdr.DataReader('DEXJPUS', 'fred', start, end)
JPY_USD = JPY_USD.resample('M').last() #日次データから月次データへ変換
JPY_USD.columns=['p'] #項目名をpに変更
```

```python
import numpy as np
JPY_USD['ln'] = np.log(JPY_USD) #対数価格
JPY_USD['r'] = JPY_USD['p'].pct_change().dropna() #収益率
JPY_USD['lndp'] = JPY_USD['ln'].diff().dropna() #対数収益率
JPY_USD['dp'] = JPY_USD['p'].diff().dropna() #価格差
```

```python
JPY_USD.tail()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig=plt.figure(figsize=(12,8))

ax=plt.subplot(2,2,1)
JPY_USD['p'].plot(color='darkblue',ax=ax)
plt.ylabel('JPY/USD')

ax2=plt.subplot(2,2,2)
JPY_USD['ln'].plot(color='darkred',ax=ax2)
plt.ylabel('ln_ JPY/USD')

ax3=plt.subplot(2,2,3)
JPY_USD['r'].plot(color='darkblue',ax=ax3)
plt.ylabel('r_ JPY/USD')

ax4=plt.subplot(2,2,4)
JPY_USD['lndp'].plot(color='darkred',ax=ax4)
plt.ylabel('lndp_ JPY/USD')
```

```python
fig=plt.figure(figsize=(12,4))

ax=plt.subplot(1,2,1)
JPY_USD['r'].hist(bins=50, color='lightgray',ax=ax)
plt.xlabel('r_ JPY/USD')
plt.ylabel('probability density function')

ax2=plt.subplot(1,2,2)
JPY_USD['lndp'].hist(bins=50, color='lightgray',ax=ax2)
plt.xlabel('lndp_ JPY/USD')
plt.ylabel('probability density function')
```

```python
print("mean %2.5f std %2.5f skew %2.5f kurt %2.5f"%(JPY_USD['r'].mean(), JPY_USD['r'].std(), JPY_USD['r'].skew(), JPY_USD['r'].kurt()))
print("mean %2.5f std %2.5f skew %2.5f kurt %2.5f"%(JPY_USD['lndp'].mean(), JPY_USD['lndp'].std(), JPY_USD['lndp'].skew(), JPY_USD['lndp'].kurt()))
```

## 日経平均

```python
start = datetime.datetime(1971,1,4)
end = datetime.datetime(2019,8,30)
N225 = pdr.DataReader('NIKKEI225', 'fred', start, end)
N225 = N225.resample('M').last() #日次データから月次データへ変換
N225.columns=['p'] #項目名をpに変更

N225['ln'] = np.log(N225) #対数価格
N225['r'] = N225['p'].pct_change().dropna() #収益率
N225['lndp'] = N225['ln'].diff().dropna() #対数収益率
N225['dp'] = N225['p'].diff().dropna() #価格差
```

```python
N225.tail()
```

```python
fig=plt.figure(figsize=(12,8))

ax=plt.subplot(2,2,1)
N225['p'].plot(color='darkblue',ax=ax)
plt.ylabel('NIKKEI225')

ax2=plt.subplot(2,2,2)
N225['ln'].plot(color='darkred',ax=ax2)
plt.ylabel('ln_ NIKKEI225')

ax3=plt.subplot(2,2,3)
N225['r'].plot(color='darkblue',ax=ax3)
plt.ylabel('r_ NIKKEI225')

ax4=plt.subplot(2,2,4)
N225['lndp'].plot(color='darkred',ax=ax4)
plt.ylabel('lndp_ NIKKEI225')
```

```python
fig=plt.figure(figsize=(12,4))

ax=plt.subplot(1,2,1)
N225['r'].hist(bins=50, color='lightgray',ax=ax)
plt.xlabel('r_ NIKKEI225')
plt.ylabel('probability density function')

ax2=plt.subplot(1,2,2)
N225['lndp'].hist(bins=50, color='lightgray',normed=True,ax=ax2)
plt.xlabel('lndp_ NIKKEI225')
plt.ylabel('probability density function')
```

```python
print("mean %2.5f std %2.5f skew %2.5f kurt %2.5f"%(N225['r'].mean(), N225['r'].std(), N225['r'].skew(), N225['r'].kurt()))
print("mean %2.5f std %2.5f skew %2.5f kurt %2.5f"%(N225['lndp'].mean(), N225['lndp'].std(), N225['lndp'].skew(), N225['lndp'].kurt()))
```
